```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from datetime import datetime
import sqlite3

DB_FILE_PATH = 'leverage.db'
FPI_TABLE_NAME = 'fpi'

try:
    url = 'https://espn.com/nfl/fpi/_/season/2020'
    page = requests.get(url)
    team_to_posteam = {'Kansas City Chiefs':'KC',
    'Tampa Bay Buccaneers':'TB',
    'New Orleans Saints':'NO',
    'Green Bay Packers':'GB',
    'Baltimore Ravens':'BAL',
    'Buffalo Bills':'BUF',
    'Los Angeles Rams':'LA',
    'Seattle Seahawks':'SEA',
    'Pittsburgh Steelers':'PIT',
    'Tennessee Titans':'TEN',
    'Indianapolis Colts':'IND',
    'San Francisco 49ers':'SF',
    'Arizona Cardinals':'ARI',
    'Atlanta Falcons':'ATL',
    'Minnesota Vikings':'MIN',
    'Cleveland Browns':'CLE',
    'New England Patriots':'NE',
    'Chicago Bears':'CHI',
    'Las Vegas Raiders':'LV',
    'Washington':'WAS',
    'Miami Dolphins':'MIA',
    'Carolina Panthers':'CAR',
    'Houston Texans':'HOU',
    'Denver Broncos':'DEN',
    'Los Angeles Chargers':'LAC',
    'New York Giants':'NYG',
    'Cincinnati Bengals':'CIN',
    'Dallas Cowboys':'DAL',
    'Philadelphia Eagles':'PHI',
    'Detroit Lions':'DET',
    'New York Jets':'NYJ',
    'Jacksonville Jaguars':'JAX'}
    soup = BeautifulSoup(page.text,'html.parser')
    tables = soup.find_all('table')
    teams = [img['alt'] for img in tables[0].find_all('img')]
    table = pd.read_html(str(tables[1]))[0]['POWER INDEX'][['FPI','OFF','DEF','ST']]
    table['team'] = pd.Series(teams).map(team_to_posteam)
    table['created_at'] = datetime.now()
    # TODO: data validations
    print(table)
    conn = sqlite3.connect(DB_FILE_PATH)
    table.to_sql(name=FPI_TABLE_NAME, con=conn, if_exists='append', index=False)
    conn.close()
except Exception as e:
    # TODO: set up emailing in case of exception (once this is a cronjob)
    print(e)
```
